```python
import requests
import pandas as pd

# URL of the Divar API for rent apartments in Tehran
url = "https://api.divar.ir/v8/web-search/tehran/rent-apartment "

# Send a GET request to the API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Convert response to JSON

    # Extracting the list of ads
    posts = data.get('widget_list', [])

    titles = []
    areas = []
    rooms = []
    prices_deposit = []
    prices_rent = []

    # Loop through each post and extract info
    for post in posts:
        data_layer = post.get('data', {})
        
        title = data_layer.get('title')
        metas = data_layer.get('metas', {})

        area = metas.get('area')
        room = metas.get('room_count')
        deposit = metas.get('deposit')
        rent = metas.get('rent')

        titles.append(title)
        areas.append(area)
        rooms.append(room)
        prices_deposit.append(deposit)
        prices_rent.append(rent)

    # Create DataFrame
    df_api = pd.DataFrame({
        'عنوان': titles,
        'مساحت': areas,
        'اتاق': rooms,
        'ودیعه': prices_deposit,
        'اجاره': prices_rent
    })

    print("✅ تعداد آگهی‌های استخراج شده:", len(df_api))
    display(df_api.head())

else:
    print("❌ درخواست ناموفق. کد وضعیت:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import json
import re
import pandas as pd

# URL صفحه
url = "https://divar.ir/s/tehran/rent-apartment "

# درخواست به سرور
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# پیدا کردن بلاک JSON
script_tag = soup.find('script', {'id': '__NEXT_DATA__'})
data = json.loads(script_tag.string)

# مسیر داده‌های آگهی
posts = data['props']['pageProps']['webAdList']['listData']

titles = []
areas = []
prices = []

for post in posts:
    title = post.get("title")
    description = post.get("description")
    
    titles.append(title)
    
    # استخراج مساحت
    area_match = re.search(r'(\d+)[\s\/]*متر', description) if description else None
    areas.append(int(area_match.group(1)) if area_match else None)
    
    # استخراج قیمت (مثال: "ودیعه: ۲,۵۰۰,۰۰۰,۰۰۰ تومان")
    price_match = re.search(r'ودیعه:\s*([\d\,]+)\s*تومان', description) if description else None
    prices.append(price_match.group(1).replace(',', '') if price_match else None)

# ساخت DataFrame
df_json = pd.DataFrame({
    'عنوان': titles,
    'مساحت': areas,
    'قیمت': prices
})

print("✅ تعداد آگهی‌های استخراج شده:", len(df_json))
display(df_json.head(10))
```

```python
import re
import json
import pandas as pd

# ۱. خواندن فایل txt
file_path = 'divar_fullrent.txt'
with open(file_path, 'r', encoding='utf-8') as f:
    html_content = f.read()

# ۲. الگوی دقیق‌تر برای پیدا کردن JSON دیوار
match = re.search(r'window\.__PRELOADED_STATE__\s*=\s*(\{.*?\});', html_content, re.DOTALL)

if match:
    try:
        # ۳. استخراج JSON و لود کردن
        raw_json = match.group(1)
        
        # ۴. تبدیل به دیکشنری پایتون
        data = json.loads(raw_json)

        # ۵. دسترسی به لیست آگهی‌ها
        posts_data = data.get('widgets', {}).get('list', [])

        # ۶. استخراج فیلدهای مهم
        posts = []
        for post in posts_data:
            if isinstance(post, dict) and post.get('widgetType') == 'POST_ROW':
                token = post.get('token')
                title = post.get('title')
                top_desc = post.get('topDescription')
                mid_desc = post.get('middleDescription')
                bot_desc = post.get('bottomDescription')

                # استخراج مساحت، ودیعه، اجاره
                area_match = re.search(r'(\d+)\s*(?:متر|متري)', title + " " + (top_desc or "") + " " + (mid_desc or ""))
                area = int(area_match.group(1)) if area_match else None

                deposit_match = re.search(r'ودیعه:\s*(\d[\d\,]*)\s*تومان', top_desc or "")
                deposit = deposit_match.group(1).replace(',', '') if deposit_match else None

                rent_match = re.search(r'اجاره:\s*(\d[\d\,]*)\s*تومان', mid_desc or "")
                rent = rent_match.group(1).replace(',', '') if rent_match else None

                # ذخیره موقعیت
                location = bot_desc

                posts.append({
                    'token': token,
                    'عنوان': title,
                    'مساحت': area,
                    'ودیعه': deposit,
                    'اجاره': rent,
                    'موقعیت': location
                })

        # ۷. ساخت DataFrame
        df_final = pd.DataFrame(posts)

        # ۸. تبدیل اعداد فارسی به انگلیسی
        def fa_to_en_numbers(text):
            if not isinstance(text, str):
                return text
            fa_digits = '۰۱۲۳۴۵۶۷۸۹'
            en_digits = '0123456789'
            translation_table = str.maketrans(fa_digits, en_digits)
            return text.translate(translation_table)

        df_final['ودیعه'] = df_final['ودیعه'].apply(fa_to_en_numbers)
        df_final['اجاره'] = df_final['اجاره'].apply(fa_to_en_numbers)

        # ۹. ذخیره به CSV
        df_final.to_csv('real_estate_data_clean.csv', index=False, encoding='utf-8-sig')
        print("✅ تعداد آگهی‌های استخراج شده:", len(df_final))
        print(df_final.head(10))

    except json.JSONDecodeError as e:
        print("❌ خطا در خواندن JSON:", e)
else:
    print("❌ JSON دیوار در فایل پیدا نشد.")
```

```python
import requests
from bs4 import BeautifulSoup
import time

base_url = "https://divar.ir/s/tehran/rent-villa/tehran-zanjan"
params = {
    "districts": "1002,1005,1006,1029,195,196,197,284,312,654,657",
    "map_bbox": "51.33375549316406,35.659141540527344,51.38261413574219,35.70193862915039",
    "map_interaction": "list_only_used",
    "map_place_hash": "1|1001,1002,1005,1006,1029,195,196,197,284,312,654,657|apartment-rent",
    "page": 1
}

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
}

listings = []

while len(listings) < 100:
    print(f"Fetching page {params['page']}...")
    response = requests.get(base_url, params=params, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    ads = soup.find_all("div", {"class": "post-card-item"})
    if not ads:
        print("No more ads found.")
        break

    for ad in ads:
        title = ad.find("div", {"class": "kt-post-card__title"}).text.strip() if ad.find("div", {"class": "kt-post-card__title"}) else "No title"
        price = ad.find("div", {"class": "kt-post-card__description"}).text.strip() if ad.find("div", {"class": "kt-post-card__description"}) else "No price"
        link = "https://divar.ir" + ad.find("a")["href"] if ad.find("a") else "No link"

        listings.append({
            "title": title,
            "price": price,
            "link": link
        })

        if len(listings) >= 100:
            break

    params["page"] += 1
    time.sleep(2)  # برای جلوگیری از بلاک شدن

print(f"Collected {len(listings)} listings.")

# ذخیره به فایل CSV اگر بخوای
import csv

with open("divar_listings.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["title", "price", "link"])
    writer.writeheader()
    writer.writerows(listings)
```

```python
import requests

def fetch_divar_listings():
    base_url = "https://api.divar.ir/v8/web-search/tehran/rent-villa"
    
    # پارامترهای لینک تو (مناطق و بقیه)
    params = {
        "districts": "1002,1005,1006,1029,195,196,197,284,312,654,657",
        "page": 1
    }
    
    all_listings = []
    
    while len(all_listings) < 100:
        print(f"Fetching page {params['page']}...")
        response = requests.get(base_url, params=params)
        if response.status_code != 200:
            print("خطا در دریافت داده‌ها")
            break
        
        data = response.json()
        
        # چک کن آیا آگهی هست یا نه
        items = data.get("widget_list", [])
        if not items:
            print("دیگه آگهی بیشتری وجود نداره")
            break
        
        # اضافه کردن آگهی‌ها به لیست
        all_listings.extend(items)
        
        # رفتن به صفحه بعد
        params["page"] += 1
    
    # فقط 100 تا آگهی برمی‌گردونیم اگر بیشتر جمع شده باشه
    return all_listings[:100]

if __name__ == "__main__":
    listings = fetch_divar_listings()
    print(f"Collected {len(listings)} listings.")
    for i, listing in enumerate(listings, 1):
        title = listing.get("title", "بدون عنوان")
        price = listing.get("data", {}).get("price", "بدون قیمت")
        print(f"{i}. {title} | قیمت: {price}")
```

```python
import requests
from bs4 import BeautifulSoup
import time

def scrape_divar_villa_listings(max_pages=5):
    base_url = "https://divar.ir/s/tehran/rent-villa/tehran-zanjan"
    params = {
        "districts": "1002,1005,1006,1029,195,196,197,284,312,654,657",
        "page": 1
    }
    all_listings = []

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
    }

    for page in range(1, max_pages + 1):
        params['page'] = page
        print(f"Fetching page {page}...")
        response = requests.get(base_url, params=params, headers=headers)
        if response.status_code != 200:
            print("خطا در دریافت صفحه")
            break
        
        soup = BeautifulSoup(response.text, "html.parser")
        ads = soup.find_all("div", {"data-testid": "listing-card"})
        if not ads:
            print("دیگه آگهی بیشتری وجود نداره")
            break
        
        for ad in ads:
            title = ad.find("h2")
            price = ad.find("div", {"data-testid": "price"})
            link = ad.find("a", href=True)
            
            all_listings.append({
                "title": title.text.strip() if title else "بدون عنوان",
                "price": price.text.strip() if price else "بدون قیمت",
                "link": "https://divar.ir" + link['href'] if link else "بدون لینک"
            })
        
        time.sleep(1)  # برای کاهش فشار به سرور و جلوگیری از بلاک شدن
    
    return all_listings

if __name__ == "__main__":
    listings = scrape_divar_villa_listings(max_pages=10)
    print(f"Collected {len(listings)} listings.")
    for i, item in enumerate(listings, 1):
        print(f"{i}. {item['title']} | {item['price']} | {item['link']}")
```
