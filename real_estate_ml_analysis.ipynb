```python
import requests
import pandas as pd

# URL of the Divar API for rent apartments in Tehran
url = "https://api.divar.ir/v8/web-search/tehran/rent-apartment "

# Send a GET request to the API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Convert response to JSON

    # Extracting the list of ads
    posts = data.get('widget_list', [])

    titles = []
    areas = []
    rooms = []
    prices_deposit = []
    prices_rent = []

    # Loop through each post and extract info
    for post in posts:
        data_layer = post.get('data', {})
        
        title = data_layer.get('title')
        metas = data_layer.get('metas', {})

        area = metas.get('area')
        room = metas.get('room_count')
        deposit = metas.get('deposit')
        rent = metas.get('rent')

        titles.append(title)
        areas.append(area)
        rooms.append(room)
        prices_deposit.append(deposit)
        prices_rent.append(rent)

    # Create DataFrame
    df_api = pd.DataFrame({
        'عنوان': titles,
        'مساحت': areas,
        'اتاق': rooms,
        'ودیعه': prices_deposit,
        'اجاره': prices_rent
    })

    print("✅ تعداد آگهی‌های استخراج شده:", len(df_api))
    display(df_api.head())

else:
    print("❌ درخواست ناموفق. کد وضعیت:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import json
import re
import pandas as pd

# URL صفحه
url = "https://divar.ir/s/tehran/rent-apartment "

# درخواست به سرور
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# پیدا کردن بلاک JSON
script_tag = soup.find('script', {'id': '__NEXT_DATA__'})
data = json.loads(script_tag.string)

# مسیر داده‌های آگهی
posts = data['props']['pageProps']['webAdList']['listData']

titles = []
areas = []
prices = []

for post in posts:
    title = post.get("title")
    description = post.get("description")
    
    titles.append(title)
    
    # استخراج مساحت
    area_match = re.search(r'(\d+)[\s\/]*متر', description) if description else None
    areas.append(int(area_match.group(1)) if area_match else None)
    
    # استخراج قیمت (مثال: "ودیعه: ۲,۵۰۰,۰۰۰,۰۰۰ تومان")
    price_match = re.search(r'ودیعه:\s*([\d\,]+)\s*تومان', description) if description else None
    prices.append(price_match.group(1).replace(',', '') if price_match else None)

# ساخت DataFrame
df_json = pd.DataFrame({
    'عنوان': titles,
    'مساحت': areas,
    'قیمت': prices
})

print("✅ تعداد آگهی‌های استخراج شده:", len(df_json))
display(df_json.head(10))
```

```python
import re
import json
import pandas as pd

# ۱. خواندن فایل txt
file_path = 'divar_fullrent.txt'
with open(file_path, 'r', encoding='utf-8') as f:
    html_content = f.read()

# ۲. الگوی دقیق‌تر برای پیدا کردن JSON دیوار
match = re.search(r'window\.__PRELOADED_STATE__\s*=\s*(\{.*?\});', html_content, re.DOTALL)

if match:
    try:
        # ۳. استخراج JSON و لود کردن
        raw_json = match.group(1)
        
        # ۴. تبدیل به دیکشنری پایتون
        data = json.loads(raw_json)

        # ۵. دسترسی به لیست آگهی‌ها
        posts_data = data.get('widgets', {}).get('list', [])

        # ۶. استخراج فیلدهای مهم
        posts = []
        for post in posts_data:
            if isinstance(post, dict) and post.get('widgetType') == 'POST_ROW':
                token = post.get('token')
                title = post.get('title')
                top_desc = post.get('topDescription')
                mid_desc = post.get('middleDescription')
                bot_desc = post.get('bottomDescription')

                # استخراج مساحت، ودیعه، اجاره
                area_match = re.search(r'(\d+)\s*(?:متر|متري)', title + " " + (top_desc or "") + " " + (mid_desc or ""))
                area = int(area_match.group(1)) if area_match else None

                deposit_match = re.search(r'ودیعه:\s*(\d[\d\,]*)\s*تومان', top_desc or "")
                deposit = deposit_match.group(1).replace(',', '') if deposit_match else None

                rent_match = re.search(r'اجاره:\s*(\d[\d\,]*)\s*تومان', mid_desc or "")
                rent = rent_match.group(1).replace(',', '') if rent_match else None

                # ذخیره موقعیت
                location = bot_desc

                posts.append({
                    'token': token,
                    'عنوان': title,
                    'مساحت': area,
                    'ودیعه': deposit,
                    'اجاره': rent,
                    'موقعیت': location
                })

        # ۷. ساخت DataFrame
        df_final = pd.DataFrame(posts)

        # ۸. تبدیل اعداد فارسی به انگلیسی
        def fa_to_en_numbers(text):
            if not isinstance(text, str):
                return text
            fa_digits = '۰۱۲۳۴۵۶۷۸۹'
            en_digits = '0123456789'
            translation_table = str.maketrans(fa_digits, en_digits)
            return text.translate(translation_table)

        df_final['ودیعه'] = df_final['ودیعه'].apply(fa_to_en_numbers)
        df_final['اجاره'] = df_final['اجاره'].apply(fa_to_en_numbers)

        # ۹. ذخیره به CSV
        df_final.to_csv('real_estate_data_clean.csv', index=False, encoding='utf-8-sig')
        print("✅ تعداد آگهی‌های استخراج شده:", len(df_final))
        print(df_final.head(10))

    except json.JSONDecodeError as e:
        print("❌ خطا در خواندن JSON:", e)
else:
    print("❌ JSON دیوار در فایل پیدا نشد.")
```
